```python
plt.plot(df["dx"], label="dx")
plt.plot(df["dy"], label="dy")
plt.legend()
plt.show()
```

```python
def competition_model(x, y, r1, r2, K1, K2, alpha, beta):
    
    dx = r1 * x * (1 - x / K1) - alpha * x * y
    
    dy = r2 * y * (1 - y / K2) - beta * x * y
    
    return dx, dy
```

```python
row = df.iloc[0]

x = row["x"]
y = row["y"]

dx, dy = competition_model(
    x, y,
    r1=0.1,
    r2=0.1,
    K1=1,
    K2=1,
    alpha=0.1,
    beta=0.1
)

dx, dy
```
