```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/os_market_data.csv")

df.head()
```

```python
df = df.sort_values("Date").reset_index(drop=True)
```

```python
df["dx"] = df["x"].diff()
df["dy"] = df["y"].diff()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.to_csv("../data/cleaned/os_model_data.csv", index=False)
```
