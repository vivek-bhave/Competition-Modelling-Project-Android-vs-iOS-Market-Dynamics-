```python
import pandas as pd
import numpy as np
import glob
```

```python
path = "../data/raw/*.csv"

files = glob.glob(path)

files
```

```python
df_list = []

for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

merged_df = pd.concat(df_list, ignore_index=True)
```

```python
merged_df.head()
merged_df.tail()
merged_df.columns
```

```python
merged_df.columns
```

```python
df = merged_df[["Date", "Android", "iOS"]]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.sort_values("Date")
```

```python
df = df[df["Date"] >= "2013-01-01"]
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv("../data/cleaned/os_market_2013_2025.csv", index=False)
```
